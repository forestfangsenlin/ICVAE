```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
MW_latent = np.load("./result/latent/MW_latent.npy")
MW_label = np.load("./result/latent/MW_label.npy")
```

### reverse the condition label from the normalized label

normalized formula: y_norm = 500*(y - np.min(y))/(np.max(y)-np.min(y))

reverse formula: **y = y_norm/500 * (np.max(y)-np.min(y))+np.min(y)**

```python
MW_label = (MW_label/500)*201.12958056818  + 150.025169192
```

```python
plt.figure(figsize=(11,8))

plt.xticks(fontsize=22)
plt.yticks(fontsize=22)

plt.xlabel('latent x', fontsize=28)
plt.ylabel('latent y', fontsize=28)


plt.scatter(MW_latent[:, 0], MW_latent[:, 1], c=MW_label[:,0], s=20, alpha=0.7,
            cmap=plt.get_cmap("jet"))

cb = plt.colorbar()
cb.set_label(label='MW',size=34, rotation='90')
cb.ax.tick_params(labelsize=24)


plt.savefig("./result/image/MW_latent.jpg", format='jpg',edgecolor='none', dpi=300)
plt.show()
```
