```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
latent_np = np.load("./result/latent/MW_latent.npy")
label_np = np.load("./result/latent/MW_label.npy")
```

### reverse the condition label from the norm label

normalized formula: y_norm = 500*(y - np.min(y))/(np.max(y)-np.min(y))

reverse formula: **y = y_norm/500 * (np.max(y)-np.min(y))+np.min(y)**

```python
latent_np = (latent_np/500)*201.12958056818  + 150.025169192
```

```python
latten_pd = pd.DataFrame({
        'x': latent_np[:, 0],
        'y': latent_np[:, 1],
        'label': label_np[:,0]})
```

```python
im = latten_pd.plot(kind="scatter", x='x', y='y', alpha=0.7, figsize=(10,7),
    c='label', cmap=plt.get_cmap("jet"), colorbar=True,
    sharex=False)

im.set_xlabel("latent x",fontsize=20)
im.set_ylabel("latent y",fontsize=20)
im.tick_params(labelsize=20)


plt.savefig("./result/image/MW_latent.jpg", format='jpg',edgecolor='none', dpi=300)
plt.show()
```
